## <b> Wirtualne środowisko (virtual environment) w Pythonie: </b>
izolowane środowisko pozwalające m.in. na pracę z różnymi wersjami oprogramowania, w tym z różnymi wersjami Pythona. 
Wszędzie zamiast "python" można wpisać nazwę dla konkretnej wersji, np. python3, python3.7 itd 
 *  Wirtualne środowisko:
     * jest izolowane od innych środowisk i dlatego mogą w nim być inne wersje tego samego pakietu
     * ma własny interpreter Pythona
     * ma własny zbiór zainstalowanych pakietów
 * <b> Tworzenie wirtualnego środowiska </b>
      * python  -m venv /path/to/virtualenv, np:
      * python -m venv ~/env1 
 * <b> Aktywacja środowiska ~/env1 </b>
      * source ~/env1/bin/activate
      * (Polecenie deactivate deaktywuje środowisko)
      * Wewnątrz środowiska pakiety instaluje się za pomocą polecenia: pip install nazwa_pakietu/biblioteki
 * <b> Deaktywacja środowiska </b>
      * deactivate      
 * <b> Usuwanie przykładowego środowiska env1</b>
      * rm -rf ~/env1
 *  <b> Dodawanie środowiska do jupyter notebook, </b>
      wewnątrz środowiska:
      * pip install ipykernel 
      * python -m ipykernel install --user --name=env1 --display-name "python env1"     
 * <b> Wybór środowiska w Jupyter notebook: </b>
      Kernel -> Change kernel -> python env1   
 * <b> Lista dostępnych wersji: </b> jupyter kernelspec list
 * <b> Usuwanie środowiska: </b>
   rm -rf ~/env1

## Podstawowe, modyfikowalne  struktury danych

1) Standardowy Python: <b> listy </b>-- jednowymiarowe pojemniki do przechowywania obiektów mieszanych typów (<b> krotki </b> to niemodyfikowalne listy)
2) Biblioteka NumPy: <b> macierze numpy (numpy arrays)</b> -- pojemniki jedno- lub wielo- wymiarowe do przechowywania obiektów tego samego typu; do szybkiego wykonywania arytmetycznych operacji; zawiera funkcje matematyczne do zastosowania na całych macierzach, bez potrzeby używania struktury pętli 
3) Biblioteka Pandas (napisana w językach C i Python): (<b> Series </b>) i <b> DataFrame </b> -- do wygodnej pracy z danymi mieszanych typów; odwoływanie się do elementów tablic za pomocą etykiet
  * (<b> Series </b> to obiekt przypominający jednowymiarową tablicę, wartości + etykiety (index)) 
  * <b> DataFrame </b> to tabela danych; w każdej kolumnie może być wartość innego typu a indeksy nie muszą być liczbami; łatwo wczytuje się nagłówki, które mogą być użyte do indeksowania. Bardzo wygodne do wczytywania plików w formacie csv. Obecnie szeroko używany standard.
4) Biblioteka Polars (napisana w języku Rust): <b> DataFrame </b> - podobnie jak w Pandas, ale inny sposób użycia - składnia funkcyjna. Szybsze!

```python
import polars as pl

# serie
s1 = pl.Series(lista2)
print(s1)
print(s1.to_numpy())  


# ramka danych podobnie jak w pandas tworzona ze słownika
dane = {
    'rok': [2020, 2021, 2022, 2023, 2024, 2025],
    'miasto': ['Gdańsk', 'Gdańsk', 'Gdańsk', 'Katowice', 'Katowice', 'Katowice'],
    'miejsce': [2, 3, 1, 1, 2, 5]
}

df1 = pl.DataFrame(dane)
print(df1)

print(df1.to_numpy())
#filtrowanie
df2=df1.filter(pl.col("rok") > 2023)
print(df2)
#dodawanie nowej kolumny
df1 = df1.with_columns(
    (pl.col("rok") + 1).alias("rok_plus_1")
)
print(df1)
```
