* wypisa na ekran nowy plik

````python
#BEGIN_SOLUTION
# function that removes columns from file. Columns to remove given as tuple of column numbers
def removeColumns(intputFileName, outputFileName, columns):
    """
    Remove columns from a file and save the result to a new file.
    
    Parameters:
    - inputFileName: str, path to the input file
    - outputFileName: str, path to the output file
    - columns: str, comma-separated string of column numbers to remove (1-indexed)
    """
    # Convert the string of columns to a list of integers
    columns = [int(col) for col in columns]
    
    with open(inputFileName, 'r') as infile, open(outputFileName, 'w') as outfile:
        for line in infile:
            # Split the line into columns
            cols = line.strip().split('\t')
            # Remove the specified columns
            cols = [col for i, col in enumerate(cols) if i + 1 not in columns]
            # Write the remaining columns to the output file
            outfile.write('\t'.join(cols) + '\n')

#END_SOLUTION
# test function removeColumns
inputFileName = '/scratch_hdd/akalinow/Zajecia/2024-2025/Lato/Narzedzia_wspierajace_programowanie/1/Narzedzia_Wspierajace_Programowanie/02_Python_obrobka_danych/people2.dat'
outputFileName = 'people2_no_columns.dat'
colums = (2,3)
removeColumns(inputFileName, outputFileName, colums)

#Print the result
#BEGIN_SOLUTION
with open(outputFileName, 'r') as f:
    for line in f:
        print(line.strip())
#END_SOLUTION
pass
````

````python
#BEGIN_SOLUTION

#Load CSV files from Narzedzia_Wspierajace_Programowanie/02_Python_obrobka_danych/Smash/runX/outdir into pandas DataFrame. 
# Keep information on the run number into pandas DataFrame.

def loadCSVIntoPandas(filePath):
    """
    Load CSV files from the given file path into a pandas DataFrame.
    
    Parameters:
    - filePath: str, path to the CSV file
    
    Returns:
    - df: pandas DataFrame containing the loaded data
    """
    import pandas as pd
    # Load the CSV file into a DataFrame
    # skip rows with comments marked by #
    df = pd.read_csv(filePath, sep=' ', header=None, comment='#', names=columns)
    
    #Extract the run number from runX part
    run_number = filePath.split('/')[-3].replace('run', '')
    # Add the run number as a new column in the DataFrame
    df['Run'] = run_number

    # Return the DataFrame
    return df

# test the function for all files in Narzedzia_Wspierajace_Programowanie/02_Python_obrobka_danych/Smash/runX directories
import os
import pandas as pd
import glob
import numpy as np


files = glob.glob('/scratch_hdd/akalinow/Zajecia/2024-2025/Lato/Narzedzia_wspierajace_programowanie/1/Narzedzia_Wspierajace_Programowanie/02_Python_obrobka_danych/Smash/run*/outdir/particle_lists.oscar')
dataframes = []
for file in files:
    df = loadCSVIntoPandas(file)
    dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
df = pd.concat(dataframes, ignore_index=True)
print(df)
#END_SOLUTION
pass
````

````python
# plot mean multiplicity of particles with charghe -1 as a function of run number
#BEGIN_SOLUTION
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Group the DataFrame by 'Run' and 'charge', and count the number of occurrences
grouped_df = df.groupby(['Run', 'charge']).size().reset_index(name='count')
# Filter for charge -1
charge_minus_1 = grouped_df[grouped_df['charge'] == -1]
# Calculate the mean multiplicity for each run
mean_multiplicity = charge_minus_1.groupby('Run')['count'].mean().reset_index()
# Plot the mean multiplicity as a function of run number
plt.plot(mean_multiplicity['Run'], mean_multiplicity['count'], marker='o')
plt.xlabel('Run Number')
plt.ylabel('Mean Multiplicity')
plt.title('Mean Multiplicity of Particles with Charge -1')
plt.grid()
#END_SOLUTION
pass
````
